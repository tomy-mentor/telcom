```python
import streamlit as st
import pickle
import pandas as pd

model= pickle.load(open('smote_rf_model', 'rb'))
features = pickle.load(open('model_features.pkl', 'rb'))
df_deploy = pickle.load(open('df_deploy.pkl', 'rb'))
```

**random customers**

```python
df_sample=df_deploy.sample(10)
```

```python
df_sample
```

```python
df = pd.get_dummies(df_sample).reindex(columns=features, fill_value=0)
```

```python
prediction = model.predict_proba(df)
```

```python
prediction[:,1]
```

```python
df_sample["Churn Prob"]=prediction[:,1]
df_sample
```

```python
df_sample.drop("Churn", axis=1, inplace=True)
```

```python
df_sample
```

```python
df_sample["Churn Prob"]=df_sample["Churn Prob"].apply(lambda x:round(x,2))
```

```python
df_sample.rename(columns={"OnlineSecurity": "Onl.Sec", "InternetService": "Service","TechSupport": "TechSupp", "MonthlyCharges": "M.Charges","TotalCharges": "T.Charges" }, inplace=True)
```
